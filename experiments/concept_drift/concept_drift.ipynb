```python
import sys
sys.path.append('../../')
import plotly.graph_objects as go 
from plotly import subplots
import plotly as py
import pandas as pd
from src.ad2s import AD2SDetector
from sklearn.metrics import roc_auc_score
import numpy as np
import time
from utils.util import cfg
```

```python
cfg.data.synthetic_ds = 3
cfg.experiments.concept_drift.ds = 3

df = pd.read_csv(cfg.data.save_path, nrows=30000)


res = []


clf = AD2SDetector()

predict = []
p_change = []
for object in df["data"]:
    p_change.append(clf.p)
    score = clf.predict(object)
    predict.append(score)


df["p"] = p_change

df = df.iloc[6500:29000]

fig = go.Figure()


fig.add_trace(
    go.Scatter(
        x=df.index,
        y=df["p"],
        marker_color='#e97b5f',
        marker_symbol="square",
        marker_size=15,
    )
)


fig.add_annotation(
    x=8300,
    y=0.35,
    text=r"$\lambda=0.001$",
    showarrow=False,
    arrowhead=1,
    font=dict(size=25, color="black"),
)

fig.add_annotation(
    x=15300,
    y=0.35,
    text=r"$\lambda=0.01$",
    showarrow=False,
    arrowhead=1,
    font=dict(size=25, color="black"),
)

fig.add_annotation(
    x=24300,
    y=0.35,
    text=r"$\lambda=0.001$",
    showarrow=False,
    arrowhead=1,
    font=dict(size=25, color="black"),
)

fig.add_vline(x=10000, line_width=2, line_dash="dash", line_color="green")
fig.add_vline(x=20000, line_width=2, line_dash="dash", line_color="green")


fig.update_layout(
    font_family="Times New Roman",
    plot_bgcolor="white",
    font_color="black",
    yaxis1=dict(
        title="Probbability (p)",
        titlefont=dict(size=35),
        tickfont=dict(size=20),
        showline=True,
        showgrid=False,
        linecolor="black",
        linewidth=2,
        mirror=True,
    ),
    xaxis1=dict(
        title="Index of observations (#3) ",
        titlefont=dict(size=35),
        tickfont=dict(size=20),
        showline=True,
        showgrid=False,
        linecolor="black",
        linewidth=2,
        mirror=True,
    ),
    height=400,
    width=700,
    margin=dict(l=0, r=0, b=0, t=0),
    legend=dict(
        x=0.75, y=1, bgcolor="rgba(0,0,0,0)", font=dict(size=20, color="black")
    ),
    legend_orientation="v",
    yaxis_range=[0, 0.5],
)
```
